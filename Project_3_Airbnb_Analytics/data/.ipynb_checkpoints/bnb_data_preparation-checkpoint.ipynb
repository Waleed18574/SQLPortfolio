```python
airbnb_searches = pd.read_csv('airbnb_searches.csv')
airbnb_searches.head()
```

```python
airbnb_searches.head(10)
```

```python
airbnb_searches['ds'] = airbnb_searches['ds'].astype(str)
```

```python
airbnb_searches['ds'] = airbnb_searches['ds'].str.replace(' 0:00','')
```

```python
airbnb_searches['ds'] = pd.to_datetime(airbnb_searches['ds'])
airbnb_searches['ds']  = airbnb_searches['ds'] .dt.strftime('%Y-%m-%d')
airbnb_searches['ds'] = pd.to_datetime(airbnb_searches['ds'])
airbnb_searches['ds'].head()
```

```python
airbnb_searches['ds_checkin'] = pd.to_datetime(airbnb_searches['ds_checkin'])
airbnb_searches['ds_checkin']  = airbnb_searches['ds_checkin'] .dt.strftime('%Y-%m-%d')
airbnb_searches['ds_checkin'] = pd.to_datetime(airbnb_searches['ds_checkin'])
airbnb_searches['ds_checkin'].head()
```

```python
airbnb_searches['ds_checkout'] = pd.to_datetime(airbnb_searches['ds_checkout'])
airbnb_searches['ds_checkout']  = airbnb_searches['ds_checkout'] .dt.strftime('%Y-%m-%d')
airbnb_searches['ds_checkout'] = pd.to_datetime(airbnb_searches['ds_checkout'])
airbnb_searches['ds_checkout'].head()
```

```python
airbnb_searches.to_csv('airbnb_searches.csv',index=False)
```
