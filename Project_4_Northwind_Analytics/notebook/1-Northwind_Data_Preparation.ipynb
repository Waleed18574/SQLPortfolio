## Writing Tables into CSV Files


data preparation for employees table

```python
# convert the null in the "ReportsTo" to 0
employees.loc[employees.ReportsTo.isnull(),'ReportsTo'] = 0
```

```python
# convert the "ReportsTo" feature in the employees table into integer data-type
employees.ReportsTo = employees.ReportsTo.astype(int)
# drop the "EmployeeId" features since it can be generated from the database using Serial data type
employees.drop('EmployeeID', axis=1, inplace=True)

# convert the "BirthDate" feature into text, extract the date, then convert it back to date
employees.BirthDate = employees.BirthDate.dt.strftime('%Y-%m-%d')
employees.BirthDate = pd.to_datetime(employees.BirthDate)

# convert the "HirehDate" feature into text, extract the date, then convert it back to date
employees.HireDate = employees.HireDate.dt.strftime('%Y-%m-%d')
employees.HireDate = pd.to_datetime(employees.HireDate)
```

suppliers table

```python
# drop the "SupplierId" features since it can be generated from the database using Serial data type
suppliers.drop('SupplierID', axis=1, inplace=True)
```

data preparation for orders table

```python
orders.OrderDate = orders.OrderDate.dt.strftime('%Y-%m-%d')
orders.OrderDate = pd.to_datetime(orders.OrderDate)

orders.RequiredDate = orders.RequiredDate.dt.strftime('%Y-%m-%d')
orders.RequiredDate = pd.to_datetime(orders.RequiredDate)

orders.ShippedDate = orders.ShippedDate.dt.strftime('%Y-%m-%d')
orders.ShippedDate = pd.to_datetime(orders.ShippedDate)
```

shippers

```python
shippers.drop('ShipperID', axis=1, inplace=True)
```

products

```python
products.drop('ProductID', axis=1, inplace=True)
```

categoires

```python
categories.drop('CategoryID', axis=1, inplace=True)
```

write tables into csv files

```python
customers.to_csv('../data/customers.csv',index=False)
employees.to_csv('../data/employees.csv',index=False)
suppliers.to_csv('../data/suppliers.csv',index=False)
shippers.to_csv('../data/shippers.csv',index=False)
products.to_csv('../data/products.csv',index=False)
categories.to_csv('../data/categories.csv',index=False)
orders.to_csv('../data/orders.csv',index=False)
order_details.to_csv('../data/order_details.csv',index=False)
```

## Establishing and Reading Tables

__customers Table__

```python
drop_tbL_customers = """
DROP TABLE IF EXISTS customers
"""
cur.execute(drop_tbL_customers)
conn.commit()
```

```python
make_tbl_customers = """
CREATE TABLE customers (
    Customer_ID CHAR(5),
    Company_Name VARCHAR(36),
    Contact_Name VARCHAR(24),
    Contact_Title VARCHAR(30),
    Address VARCHAR(46),
    City VARCHAR(18),
    Region VARCHAR(14),
    PostalCode VARCHAR(10),
    Country VARCHAR(14),
    Phone VARCHAR(18),
    Fax VARCHAR(18));
"""

cur.execute(make_tbl_customers)
conn.commit()
```
